```julia
function odemodel!(dydt, params, t)
    b, dm, kb, ku, f, thetar, k_cm, s0, gmax, cl, thetax, Kt, M, we, Km, vm, nx, Kq, Kp, vt, wr, wq, wp, hq, nr, ns = params
    # i_cr, i_em, i_cp, i_cq, i_ct, i_et, i_cm, i_mt, i_mm, i_q, i_p, i_si, i_mq, i_mp, i_mr, i_r, i_a = init

    cr, em, cp, cq, ct, et, cm, mt, mm, q, p, si, mq, mp, mr, r, a = zeros(length(dydt))

    Kgamma = gmax/Kp
    gamma = gmax*a/(Kgamma+a)
    ttrate = (cq + cr + cp + ct + cm)*gamma
    lam = ttrate/M
    vimp = et*vt*s0/(Kt+s0)
    nucat = em*vm*si/(Km+si)

    dydt[1] = +r*mr*kb - cr*ku - cr*lam - cr*gamma/nr - f*cr
    dydt[2] = - lam*em + cm*gamma/nx
    dydt[3] =  +r*mp*kb - cp*ku - cp*lam - cp*gamma/nx - f*cp
    dydt[4] =  +r*mq*kb - cq*ku - cq*lam - cq*gamma/nx - f*cq
    dydt[5] =  +r*mt*kb - ct*ku - ct*lam - ct*gamma/nx - f*ct
    dydt[6] = - lam*et + ct*gamma/nx
    dydt[7] =  +r*mm*kb - cm*ku - cm*lam - cm*gamma/nx - f*cm
    dydt[8] =  +we*a/(thetax+a) - mt*lam+dm - r*mt*kb + ct*ku + ct*gamma/nx
    dydt[9] =  +we*a/(thetax+a) - mm*lam+dm - r*mm*kb + cm*ku + cm*gamma/nx
    dydt[10] = - lam*q + cq*gamma/nx
    dydt[11] = - lam*p + cp*gamma/nx
    dydt[12] = - lam*si + vimp - nucat
    dydt[13] =  +wq*a/(thetax+a)/1+(q/Kq)^hq - mq*lam+dm - r*mq*kb + cq*ku + cq*gamma/nx
    dydt[14] = +wp*a/(thetax+a) - mp*lam+dm - r*mp*kb + cp*ku + cp*gamma/nx
    dydt[15] =  +wr*a/(thetar+a) - mr*lam+dm- r*mr*kb + cr*ku + cr*gamma/nr
    dydt[16] = - lam*r - r*mr*kb - r*mt*kb - r*mm*kb - r*mq*kb - r*mp*kb + cr*ku + ct*ku + cm*ku + cq*ku + cp*ku + cr*gamma/nr + cr*gamma/nr + ct*gamma/nx + cm*gamma/nx + cq*gamma/nx + cp*gamma/nx
    dydt[17] =  +ns*nucat - ttrate - lam*a
    # cr =  +r*mr*kb - cr*ku - cr*lam - cr*gamma/nr - f*cr
    # em = - lam*em + cm*gamma/nx
    # cp =  +r*mp*kb - cp*ku - cp*lam - cp*gamma/nx - f*cp
    # cq =  +r*mq*kb - cq*ku - cq*lam - cq*gamma/nx - f*cq
    # ct =  +r*mt*kb - ct*ku - ct*lam - ct*gamma/nx - f*ct
    # et = - lam*et + ct*gamma/nx
    # cm =  +r*mm*kb - cm*ku - cm*lam - cm*gamma/nx - f*cm
    # mt =  +we*a/(thetax+a) - mt*lam+dm - r*mt*kb + ct*ku + ct*gamma/nx
    # mm =  +we*a/(thetax+a) - mm*lam+dm - r*mm*kb + cm*ku + cm*gamma/nx
    # q = - lam*q + (cq*gamma)/nx
    # p = - lam*p + (cp*gamma)/nx
    # si = - lam*si + vimp - nucat
    # mq =  +wq*a/(thetax+a)/1+(q/Kq)^hq - mq*lam+dm - r*mq*kb + cq*ku + cq*gamma/nx
    # mp =  +wp*a/(thetax+a) - mp*lam+dm - r*mp*kb + cp*ku + cp*gamma/nx
    # mr =  +wr*a/(thetar+a) - mr*lam+dm- r*mr*kb + cr*ku + (cr*gamma)/nr
    # r = - lam*r - r*mr*kb - r*mt*kb - r*mm*kb - r*mq*kb - r*mp*kb + cr*ku + ct*ku + cm*ku + cq*ku + cp*ku + cr*gamma/nr + cr*gamma/nr + ct*gamma/nx + cm*gamma/nx + cq*gamma/nx + cp*gamma/nx
    # a = +ns*nucat - ttrate - lam*a

    # dydt[1] = cr
    # dydt[2] = em
    # dydt[3] =  cp
    # dydt[4] =  cq
    # dydt[5] =  ct
    # dydt[6] = et
    # dydt[7] =  cm
    # dydt[8] =  mt
    # dydt[9] =  mm
    # dydt[10] = q
    # dydt[11] = p
    # dydt[12] = si
    # dydt[13] =  mq
    # dydt[14] = mp
    # dydt[15] =  mr
    # dydt[16] = r
    # dydt[17] = a
end

# parameters
b= 0
dm= 0.1
kb= 1
ku= 1.0
f= 1
thetar= 426.8693338968694
k_cm= 0.005990373118888
s0= 1e4
gmax= 1260.0
cl= 0
thetax= 4.379733394834643
Kt= 1.0e3
M= 1.0e8
we= 4.139172187824451
Km= 1.0e3
vm= 5800.0
nx= 300.0
Kq= 1.522190403737490e+05
Kp= 180.1378030928276
vt= 726.0
wr= 929.9678874564831
wq= 948.9349882947897
wp= 0.0
hq= 4
nr= 7549.0
ns= 0.5
params= [b, dm, kb, ku, f, thetar, k_cm, s0, gmax, cl, thetax, Kt, M, we, Km, vm, nx, Kq, Kp, vt, wr, wq, wp, hq, nr, ns]

cr_0= 0
em_0= 0
cp_0= 0
cq_0= 0
ct_0= 0
et_0= 0
cm_0= 0
mt_0= 0
mm_0= 0
q_0= 0
p_0= 0
si_0= 0
mq_0= 0
mp_0= 0
mr_0= 0
r_0= 10.0
a_0= 1000.0

init= [cr_0, em_0, cp_0, cq_0, ct_0, et_0, cm_0, mt_0, mm_0, q_0, p_0, si_0, mq_0, mp_0, mr_0, r_0, a_0]
```
