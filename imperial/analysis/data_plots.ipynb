```python
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_colD = pd.read_csv('/content/drive/MyDrive/data/bo_data/colD_supf2_bo.csv')
df_f4c = pd.read_csv('/content/drive/MyDrive/data/bo_data/fig4c_bo.csv')
df_f4e = pd.read_csv('/content/drive/MyDrive/data/bo_data/fig4e_rtcoff_bo.csv')
df_f4f = pd.read_csv('/content/drive/MyDrive/data/bo_data/fig4f_rtcon_bo.csv')
```

```python
df_colD.set_index('Time (hours)', drop=False, inplace=True)
df_colD = df_colD.drop('Time (hours)', axis=1)

df_f4c.set_index('Time (hpi)', drop=False, inplace=True)
df_f4c = df_f4c.drop('Time (hpi)', axis=1)

df_f4e.set_index('Time (hpi)', drop=False, inplace=True)
df_f4e = df_f4e.drop('Time (hpi)', axis=1)

df_f4f.set_index('Time (hpi)', drop=False, inplace=True)
df_f4f = df_f4f.drop('Time (hpi)', axis=1)
```

```python
df_f4c[['WT','HPX']].plot(kind='bar', yerr=[df_f4c['WT_std'],df_f4c['HPX_std']], rot=0, ylabel='Miller units', title='WT vs Hpx').legend(loc='upper left')
```

```python
df_f4e[['WT','HPX']].plot(yerr=[df_f4e['WT_std'],df_f4e['HPX_std']], rot=0, ylabel='OD600', fmt='o-', title='WT vs RtcOFF').legend(loc='best')
```

```python
df_f4f[['WT','HPX']].plot(yerr=[df_f4f['WT_std'],df_f4f['HPX_std']], rot=0, ylabel='OD600', fmt='o-', title='WT vs RtcON').legend(loc='best')
```

```python
plt.subplots(4,2, figsize=(10,15))
plt.subplots_adjust(hspace=.4)
leg = (['no Colicin D', 'with Colicin D'])

plt.subplot(4,2,1)
df_colD[['0nM, WT', '250nM, WT']].plot(yerr=[df_colD['0nM, WT std'], df_colD['250nM, WT std']], ylabel='Growth (OD600)', title='WT', ax=plt.gca()).legend(leg)

plt.subplot(4,2,2)
df_colD[['0nM, nA', '250nM, nA']].plot(yerr=[df_colD['0nM, nA std'], df_colD['250nM, nA std']], ylabel='Growth (OD600)', title='rtcA/pBAD18cm', ax=plt.gca()).legend(leg)

plt.subplot(4,2,3)
df_colD[['0nM, nA_A', '250nM, nA_A']].plot(yerr=[df_colD['0nM, nA_A std'], df_colD['250nM, nA_A std']], ylabel='Growth (OD600)', title='rtcA/pBAD18cm(rtcA)', ax=plt.gca()).legend(leg)

plt.subplot(4,2,4)
df_colD[['0nM, nB', '250nM, nB']].plot(yerr=[df_colD['0nM, nB std'], df_colD['250nM, nB std']], ylabel='Growth (OD600)', title='rtcB/pBAD18cm', ax=plt.gca()).legend(leg)

plt.subplot(4,2,5)
df_colD[['0nM, nB_B', '250nM, nB_B']].plot(yerr=[df_colD['0nM, nB_B std'], df_colD['250nM, nB_B std']], ylabel='Growth (OD600)', title='rtcB/pBAD18cm(rtcB)', ax=plt.gca()).legend(leg)

plt.subplot(4,2,6)
df_colD[['0nM, nB_Bmut', '250nM, nB_Bmut']].plot(yerr=[df_colD['0nM, nB_Bmut std'], df_colD['250nM, nB_Bmut std']], ylabel='Growth (OD600)', title='rtcB/pBAD18cm(rtcB mut)', ax=plt.gca()).legend(leg)

plt.subplot(4,2,7)
df_colD[['0nM, nR', '250nM, nR']].plot(yerr=[df_colD['0nM, nR std'], df_colD['250nM, nR std']], ylabel='Growth (OD600)', title='rtcR/pBAD18cm', ax=plt.gca()).legend(leg)

plt.subplot(4,2,8)
df_colD[['0nM, nR_R', '250nM, nR_R']].plot(yerr=[df_colD['0nM, nR_R std'], df_colD['250nM, nR_R std']], ylabel='Growth (OD600)', title='rtcR/pBAD18cm(rtcR)', ax=plt.gca()).legend(leg)

```
