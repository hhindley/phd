```python
std = [[df.iloc[i].loc['0nM', 'WT'].std() for i in range(len(df.index))], [df.iloc[i].loc['250nM', 'WT'].std() for i in range(len(df.index))], [df.iloc[i].loc['0nM','nA'].std() for i in range(len(df.index))], [df.iloc[i].loc['250nM','nA'].std() for i in range(len(df.index))], [df.iloc[i].loc['0nM','nA_A'].std() for i in range(len(df.index))], [df.iloc[i].loc['250nM','nA'].std() for i in range(len(df.index))],[df.iloc[i].loc['0nM','nB'].std() for i in range(len(df.index))], [df.iloc[i].loc['250nM','nB'].std() for i in range(len(df.index))], [df.iloc[i].loc['0nM','nB_B'].std() for i in range(len(df.index))], [df.iloc[i].loc['250nM','nB_B'].std() for i in range(len(df.index))], [df.iloc[i].loc['0nM','nB_Bmut'].std() for i in range(len(df.index))], [df.iloc[i].loc['250nM','nB_Bmut'].std() for i in range(len(df.index))], [df.iloc[i].loc['0nM','nR'].std() for i in range(len(df.index))], [df.iloc[i].loc['250nM','nR'].std() for i in range(len(df.index))], [df.iloc[i].loc['0nM','nR_R'].std() for i in range(len(df.index))], [df.iloc[i].loc['250nM','nR_R'].std() for i in range(len(df.index))]]
df_std = pd.DataFrame({'0nM, WT std':pd.Series(std[0]), '250nM, WT std':pd.Series(std[1]), '0nM, nA std':pd.Series(std[2]), '250nM, nA std':pd.Series(std[3]), '0nM, nA_A std':pd.Series(std[4]), '250nM, nA_A std':pd.Series(std[5]), '0nM, nB std':pd.Series(std[6]), '250nM, nB std':pd.Series(std[7]), '0nM, nB_B std':pd.Series(std[8]), '250nM, nB_B std':pd.Series(std[9]), '0nM, nB_Bmut std':pd.Series(std[10]), '250nM, nB_Bmut std':pd.Series(std[11]), '0nM, nR std':pd.Series(std[12]), '250nM, nR std':pd.Series(std[13]), '0nM, nR_R std':pd.Series(std[14]), '250nM, nR_R std':pd.Series(std[15])})
```

```python
data = pd.concat([df_means, df_std], axis=1)
data = data.rename(columns={'Time':'Time (hours)'})
data['Time (hours)'] = data['Time (hours)']/60
data.set_index('Time (hours)', drop=False, inplace=True)
data = data.drop('Time (hours)', axis=1)
data
```
